# Matplotlib Basics
<br>Functional Way

````python
import numpy as np
import matplotlib.pyplot as plt
````


If it don't show the visualizations, try the magic line: 
```python
%matplotlib inline
```

````python
x = np.arange(0, 10)
y = x ** 2
````

**basic plot**

````python
plt.plot(x, y)
# for python scripts
# plt.show()
````

For removing the text output from the matplotlib object, you can:
- use ; after plotting expression
- use plt.show()

````python
plt.scatter(x, y);
````

````python
plt.scatter(x, y)
plt.show()
````

**adding title and labels**

````python
plt.plot(x, y)
plt.xlabel('my x label')
plt.ylabel('my y label')
plt.title('My Viz')
````

**limit the x and y axis**

````python
plt.plot(x, y)
plt.xlabel('my x label')
plt.ylabel('my y label')
plt.xlim(0, 6)
plt.ylim(0, 30)
plt.title('My Viz')
````

**save the figure**

````python
plt.plot(x, y)
plt.xlabel('my x label')
plt.ylabel('my y label')
plt.xlim(0, 6)
plt.ylim(0, 30)
plt.title('My Viz')
plt.savefig('myfirstplot.png')
````
