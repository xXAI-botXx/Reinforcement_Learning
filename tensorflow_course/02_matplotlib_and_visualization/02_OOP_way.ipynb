# Matplotlib with OOP way

```python
import numpy as np
import matplotlib.pyplot as plt
```

data

```python
a = np.linspace(0, 10, 11)
b = a**4
```

```python
x = np.arange(0, 10)
y = x * 2
```

**basic plot figure OOP**

```python
fig = plt.figure()
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y)
plt.show()
```

**more complex example**

```python
fig = plt.figure()

# axes 1:
ax1 = fig.add_axes([0, 0, 1, 1])
ax1.plot(a, b)

ax1.set_xlim(0, a.max())
ax1.set_ylim(0, b.max())
ax1.set_title('Potential growth')
ax1.set_xlabel('normal numbers')
ax1.set_ylabel('big numbers')

# axes 2

# relative to figure size:
#                    x    y  width heigth
ax2 = fig.add_axes([0.1, 0.4, 0.5, 0.5])
ax2.plot(x, y)

ax2.set_xlim(0, x.max())
ax2.set_ylim(0, y.max())
ax2.set_title('linear growth')
ax2.set_xlabel('normal numbers')
ax2.set_ylabel('big numbers')

 # show
plt.show()
```

**nice zoom in example**

```python
fig = plt.figure()

# axes 1:
ax1 = fig.add_axes([0, 0, 1, 1])
ax1.plot(a, b)

ax1.set_xlim(0, a.max())
ax1.set_ylim(0, b.max())
ax1.set_title('Power 4')
ax1.set_xlabel('normal numbers')
ax1.set_ylabel('big numbers')

# axes 2

# relative to figure size:
#                    x    y  width heigth
ax2 = fig.add_axes([0.15, 0.4, 0.5, 0.5])
ax2.plot(a, b)

ax2.set_xlim(1, 3)
ax2.set_ylim(0, 100)
ax2.set_title('Zoom in')
ax2.set_xlabel('normal numbers')
ax2.set_ylabel('big numbers')

 # show
plt.show()
```

**figure parameters**

```python
# dpi = quality of the figure
# figsize -> width, height in inch
fig = plt.figure(figsize=(8, 2.5), dpi=150)

ax1 = fig.add_axes([0, 0 ,1, 1])
ax1.plot(a, b)

# bbox_inches for make sure the axes label
fig.savefig('my_first_oop_fig.png', bbox_inches='tight')

plt.show()
```
