# Matplotlib Subplots

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0, 10, 11)
b = a ** 4
```

```python
x = np.arange(0, 10)
y = x * 2
```

**subplot basic**

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(a, b)

axes[1].plot(x, y)

plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1)

axes[0].plot(a, b)

axes[1].plot(x, y)

plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(a, b)

axes[1][1].plot(x, y)

plt.show()
```

**advanced example**

```python
fig, axes = plt.subplots(nrows=1, ncols=2, sharey=True)

axes[0].plot(a, b)

axes[1].plot(x, y)

axes_1_2 = fig.add_axes([0.6, 0.4, 0.25, 0.25])
axes_1_2.plot(x, y)
axes_1_2.set_xlim(2, 6)
axes_1_2.set_title("Zoomed in")

plt.show()
```

**loop through axes + tight layout**

```python
a_2 = np.linspace(0, 1000, 1001)
b_2 = a_2 ** 4
```

```python
fig, axes = plt.subplots(nrows=10, ncols=4, figsize=(10, 15))

i = 1
for cur_axes_arr in axes:
    for cur_axes in cur_axes_arr:
        cur_axes.plot(a_2[i:i+5], b_2[i:i+5])
        i = i + 5

plt.tight_layout()
plt.show()
```

**adjust**

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 6))

axes[0][0].plot(a,b)

axes[0][1].plot(x,y)

axes[1][0].plot(x,y)
axes[1][0].set_xlabel("fantastic label")

axes[1][1].plot(a,b)
axes[1][1].set_title("Wonderful title")
axes[1][1].set_xlim(0, 5)

fig.subplots_adjust(wspace=1.0, hspace=0.5)
fig.suptitle("Figure title", fontsize=20)

fig.savefig('my_first_subplot_figure', bbox_inches='tight')

plt.show()
```
