# Matplotlib Styling

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 11, 10)
```

**legend**

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax1.plot(x, x*2, label='x*2')
ax1.plot(x, x**2, label='x^2')
#ax1.legend(loc='center left')
#ax1.legend(loc='upper center')
ax1.legend(loc=(1.02, 0.85))

# 'best'            0
# 'upper right'     1
# 'upper left'      2
# 'lower left'      3
# 'lower right'     4
# 'right'           5
# 'center left'     6
# 'center right'    7
# 'lower center'    8
# 'upper center'    9
# 'center'          10

plt.show()
```

**color**

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, x, color='#6725a8', label='x v x')
ax.plot(x-1, x -2, color='#f7ce00', label='x-1 v x-2')
ax.legend()

plt.show()
```

**line width**<br>
lw & linewidth

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, x, color='#6725a8', label='x v x', lw=10)
ax.plot(x-1, x -2, color='#f7ce00', label='x-1 v x-2', linewidth=2.4)
ax.legend()

plt.show()
```

**line styles**<br>
ls & linestyle
<br>
possible options: '--', '-', ':', '-.', 'steps'

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, x, color='#6725a8', label='x v x', ls='--')
ax.plot(x-1, x -2, color='#f7ce00', label='x-1 v x-2', lw=4, linestyle=':')
ax.legend()

plt.show()
```

Alternative:

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
lines = ax.plot(x, x, color='#6725a8', label='x v x', lw=5)
# small line + small space + bigger line + small space
print(lines)
lines[0].set_dashes([5,2,10,2])

plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
lines = ax.plot(x, x, color='#6725a8', label='x v x', lw=5)
lines[0].set_dashes([1,1,3,1,5,1,7,1,5,1,3,1,1])

plt.show()
```

**markers**<br>
[show all options](https://matplotlib.org/stable/api/markers_api.html)

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, x, color='#6725a8', label='x v x', lw=5, marker='o', markersize=20)
ax.plot(x, x-5, color='#6725a8', label='x v x-5', lw=3, marker='v', markersize=15)

plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, x, color='#6725a8', label='x v x', lw=5, marker='o', markersize=20, 
        markerfacecolor='red', markeredgewidth=8, markeredgecolor='orange')
ax.plot(x, x-5, color='#6725a8', label='x v x-5', lw=3, marker='v', markersize=15,
       markerfacecolor='red', markeredgewidth=3, markeredgecolor='orange')

plt.show()
```

**tick positioning**

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, x, color='#6725a8', label='x v x', lw=5, marker='o', markersize=20, 
        markerfacecolor='red', markeredgewidth=8, markeredgecolor='orange')
ax.plot(x, x-5, color='#6725a8', label='x v x-5', lw=3, marker='v', markersize=15,
       markerfacecolor='red', markeredgewidth=3, markeredgecolor='orange')

ax.set_yticks([np.random.randint(ax.get_yticks().min(), ax.get_yticks().max()) for i in range(len(ax.get_yticks()))])

plt.show()
```
