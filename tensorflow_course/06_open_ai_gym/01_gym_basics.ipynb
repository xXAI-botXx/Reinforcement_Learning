**choose randoma actions**

```python
env.reset()
```

```python
images = []

for step in range(200):
    images += [env.render(mode='rgb_array')]

    # grab a possible action
    random_action = env.action_space.sample()
    #print(random_action)

    # use action
    observation, reward, done, info = env.step(random_action)

    print(f"Reward = {reward}")
    print(f"Done = {done}")
    print(f"Info = {info}")
    print("-"*12)

    if done:
        break

    time.sleep(0.1)

env.close()
```

```python
%matplotlib inline

for image in images:
    plt.imshow(image)
    plt.show()
```
