```python
import gym
```

```python
env_name = 'MountainCar-v0'
env = gym.make(env_name)
```

```python
env.seed(42)    # sets the start position to the same
observation = env.reset()

for step in range(600):
    env.render(mode='human')
    action = simple_agent(observation)
    observation, reward, done, info = env.step(action)

    print(f"Velocity = {observation[0]}")
    print(f"Position = {observation[1]}")
    #print(f"Reward = {reward}")
    #print(f"Done = {done}")
    #print(f"Info = {info}")
    print("-"*12)

    time.sleep(0.01)

env.close()
```

```python
def simple_agent(observation):
    # OBSERVATION
    position, velocity = observation

    # When to go right
    if 0.4 > position > -0.1:
        action = 2

    # When to go left
    elif velocity < 0 and position < -0.2:
        action = 0

    # when to not to do anything
    else:
        action = 1

    # ACTION
    return action
```
